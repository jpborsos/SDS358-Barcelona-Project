```python
def check_if_turnover(event):
    turnover = False
#    if event['type']['name'] == ['50-50'] and event['50-50']['counterpress'] == True:
#        turnover = True
    if event['type']['name'] == ['Ball Recovery']:
        turnover = True
    elif event['type']['name'] == 'Duel' and event['duel']['type'] == 'Tackle' and event['duel']['outcome']['name'] == 'Won':
        turnover = True
    elif event['type']['name'] == 'Duel' and event['duel']['type'] == 'Tackle' and event['duel']['outcome']['name'] == 'Success':
        turnover = True
    elif event['type']['name'] == 'Duel' and event['duel']['type'] == 'Tackle' and event['duel']['outcome']['name'] == 'Success In Play':
        turnover = True
    elif event['type']['name'] == 'Duel' and event['duel']['type'] == 'Tackle' and event['duel']['outcome']['name'] == 'Success Out':
        turnover = True
    elif event['type']['name'] == 'Interception' and event['interception']['outcome']['name'] == 'Won':
        turnover = True
    elif event['type']['name'] == 'Interception' and event['interception']['outcome']['name'] == 'Success':
        turnover = True
    elif event['type']['name'] == 'Interception' and event['interception']['outcome']['name'] == 'Success In Play':
        turnover = True
    elif event['type']['name'] == 'Interception' and event['interception']['outcome']['name'] == 'Success Out':
        turnover = True
    elif event['type']['name'] == 'Pass' and event['pass'].get('type') and event['pass']['type']['name'] == 'Recovery':
        turnover = True
    elif event['type']['name'] == 'Pass' and event['pass'].get('type') and event['pass']['type']['name'] == 'Interception':
        turnover = True


    return turnover
    
```
